```
def create_correlation(initial_df):
  #creating the correlation between the stock code and the description
  corr = []
  for i in initial_df.StockCode.unique():
    corr.append({'StockCode': i, 'Description': initial_df.loc[df['StockCode'] == i, 'Description'].iloc[0]})
  return corr
```

```
import time
from datetime import datetime, timedelta
from random import randrange

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)

def str_time_prop(start):

    stime = datetime.strptime(start, '%m/%d/%Y %H:%M')
    etime = datetime.strptime(start, '%m/%d/%Y %H:%M') + timedelta(hours=1)

    new_date = random_date(stime, etime)
    return new_date


def generate_random_date(start):
    return str_time_prop(start)
```
