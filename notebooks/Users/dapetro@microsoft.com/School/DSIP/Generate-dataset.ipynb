```
def create_correlation(initial_df):
  #creating the correlation between the stock code and the description
  corr = []
  for i in initial_df.StockCode.unique():
    corr.append({'StockCode': i, 'Description': initial_df.loc[df['StockCode'] == i, 'Description'].iloc[0]})
  return corr
```

```
import time
from datetime import datetime
from datetime import timedelta

def str_time_prop(start, end, format, prop):
    """Get a time at a proportion of a range of two formatted times.

    start and end should be strings specifying times formated in the
    given format (strftime-style), giving an interval [start, end].
    prop specifies how a proportion of the interval to be taken after
    start.  The returned time will be in the specified format.
    """

    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(start, format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(format, time.localtime(ptime))


def random_date(start, end, prop):
    return str_time_prop(start, end, '%m/%d/%Y %I:%M', prop)
```

```
def random_channel(row):
    channel = ''
    rnd =  random.randrange(101)
    if  0 < rnd <= 10:
        channel = 'pos'
    elif 10 < rnd <= 50:
        channel = 'online'
    elif 51 < rnd <= 100:
        channel = 'virtual'
    else:
        channel = 'mobile'
    
    return channel
```

```
import string
def paymentMethod_cardNumber(row):
    card_number = ''
    numbers = string.digits
    for part in range(4):
        card_number += ''.join((random.choice(numbers) for i in range(4))) + '-'
    return card_number[:-1]
```

```
ml_set['paymentMethod_cardNumber'] = ml_set.apply(paymentMethod_cardNumber, axis = 1)
```
